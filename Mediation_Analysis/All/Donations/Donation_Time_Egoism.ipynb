## Mall Donation Money Spendentafel

```python
# Mall Data Set

m_d_m_df = df[df["Experiment_Type"] == "M_D_M"].copy()
m_d_s_m_df = df[df["Experiment_Type"] == "M_D_S_M"].copy()


columnNames = {
    'Visibility_0_Private_1_Public': 'Spendentafel',
    'Style_Glam0_Hippie1': 'StyleGlam0Hippie1'
}

m_d_m_df  = m_d_m_df.rename(columns=columnNames)
m_d_s_m_df = m_d_s_m_df.rename(columns=columnNames)

vars_needed = [
    'StyleGlam0Hippie1',
    'DonationMoney',
    'Egoism'
    'Spendentafel'
]

mall_dm = pd.concat([m_d_m_df, m_d_s_m_df], ignore_index=True)

# Inspect raw values BEFORE dropna()
print("=== Raw value counts BEFORE dropna ===\n")

for col in vars_needed:
    print(f"\n--- {col} ---")
    print(mall_dm[col].value_counts(dropna=False))

print("\n=== Missing values BEFORE dropna ===")
print(mall_dm[vars_needed].isna().sum())
print("\n=== Total rows BEFORE dropna ===")
print(len(mall_dm))


mall_donation_money_spendentafel = mall_dm[vars_needed].dropna().copy()

mall_donation_money_spendentafel.columns.tolist()


```
