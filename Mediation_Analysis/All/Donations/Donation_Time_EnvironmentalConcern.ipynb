## Mall Donation Time Spendentafel

```python
# Mall Data Set

m_d_t_df = df[df["Experiment_Type"] == "M_D_T"].copy()
m_d_s_t_df = df[df["Experiment_Type"] == "M_D_S_T"].copy()


columnNames = {
    'Visibility_0_Private_1_Public': 'Spendentafel',
    'Style_Hippie0_Glam1': 'StyleGlam1Hippie0',
    'Green_Values' : 'GreenValues',
    'Environmental_Concern' : 'EnvironmentalConcern'

}

m_d_t_df  = m_d_t_df.rename(columns=columnNames)
m_d_s_t_df = m_d_s_t_df.rename(columns=columnNames)

vars_needed = [
    'StyleGlam1Hippie0',
    'DonationTime',
    'Spendentafel',
    'EnvironmentalConcern'
]

mall_dm = pd.concat([m_d_s_t_df, m_d_t_df], ignore_index=True)

# Inspect raw values BEFORE dropna()
print("=== Raw value counts BEFORE dropna ===\n")

for col in vars_needed:
    print(f"\n--- {col} ---")
    print(mall_dm[col].value_counts(dropna=False))

print("\n=== Missing values BEFORE dropna ===")
print(mall_dm[vars_needed].isna().sum())
print("\n=== Total rows BEFORE dropna ===")
print(len(mall_dm))


mall_donation_time_spendentafel = mall_dm[vars_needed].dropna().copy()

mall_donation_time_spendentafel.columns.tolist()


```
